```python
import pandas as pd
import numpy as np
```

```python
# read the wiki html directly into  using pandas read_html. this reads 2 tables present in the html into and gives a list of 2 different dataframes
df = pd.read_html(r'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header =None)
# we need only the first table (dataframe). read the 1st occurance of the list and it gives us the required dataframe
df = df[0]
df.head()
```

```python
# Get names of indexes for which column Borough has a value of Not assigned
indexNames = df[df['Borough'] == 'Not assigned'].index

print(indexNames)
 
# Delete these row indexes from dataFrame
df.drop(indexNames, inplace=True)

df
```

```python
df1=df.groupby(['Postcode','Borough'], as_index=False,sort=False).agg( ','.join)
df1.Neighbourhood[df1['Neighbourhood'] == 'Not assigned'] = df1.Borough
df1

                       
```
